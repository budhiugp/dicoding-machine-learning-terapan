```python
# Define column names
COL_APP   = "name"
COL_GENRE = "tags"
COL_DESC  = "sinopsis"
COL_CAT   = "Category"
COL_RATING = "rankScore"
COL_ISRANKED = "rank type"
COL_RANK = "rank"
COL_PRICE = "price"
COL_PUBLISHER = "Published by"


# Check existing columns
games_df_2.columns
```

Standardized column mappings to ensure consistent access to key features in the dataset, such as app name, genre, description, category, rating, etc. Since different datasets may use different column names, these variables allow the code to remain flexible and reusable

```python
# Prepare the dataset
import re
import numpy as np
import pandas as pd

# -------------------------
# Helpers
# -------------------------
def clean_text(s):
    """Lowercase + normalize whitespace. NaN -> ''."""
    if pd.isna(s):
        return ""
    s = str(s).lower()
    s = re.sub(r"\s+", " ", s)
    return s.strip()

def parse_rank(s):
    """
    Parse '16 / 4512' -> (16, 4512).
    Returns (np.nan, np.nan) if can't parse.
    """
    if pd.isna(s):
        return (np.nan, np.nan)
    m = re.match(r"^\s*(\d+)\s*/\s*(\d+)\s*$", str(s))
    if not m:
        return (np.nan, np.nan)
    return (int(m.group(1)), int(m.group(2)))

def price_bucket(p):
    """Bucket price for content-based features."""
    if pd.isna(p):
        return "unknown_price"
    try:
        p = float(p)
    except Exception:
        return "unknown_price"
    if p == 0:
        return "free"
    elif p < 10:
        return "budget"
    elif p < 30:
        return "mid"
    else:
        return "premium"

games_df_3 = games_df_2.copy()

# -------------------------
# Validate columns
# -------------------------
required_cols = [COL_APP, COL_GENRE]
missing = [c for c in required_cols if c not in games_df_3.columns]
if missing:
    raise ValueError(f"Missing required columns: {missing}")

# Ensure optional columns exist
optional_cols = [COL_DESC, COL_CAT, COL_PUBLISHER, COL_ISRANKED, COL_RANK, COL_PRICE, COL_RATING]
for c in optional_cols:
    if c not in games_df_3.columns:
        games_df_3[c] = np.nan

# -------------------------
# Basic cleanup / typing
# -------------------------
games_df_3[COL_APP] = games_df_3[COL_APP].astype(str).str.strip()
games_df_3[COL_GENRE] = games_df_3[COL_GENRE].fillna("").astype(str)
games_df_3[COL_DESC] = games_df_3[COL_DESC].fillna("").astype(str)
games_df_3[COL_CAT] = games_df_3[COL_CAT].fillna("").astype(str)
games_df_3[COL_PUBLISHER] = games_df_3[COL_PUBLISHER].fillna("").astype(str)
games_df_3[COL_ISRANKED] = games_df_3[COL_ISRANKED].fillna("").astype(str)
games_df_3[COL_RANK] = games_df_3[COL_RANK].fillna("").astype(str)

# Make numeric where appropriate
games_df_3[COL_PRICE] = pd.to_numeric(games_df_3[COL_PRICE], errors="coerce")
games_df_3[COL_RATING] = pd.to_numeric(games_df_3[COL_RATING], errors="coerce")
games_df_3[COL_RATING] = games_df_3[COL_RATING].fillna(0)

# Remove duplicates by game name
games_df_3 = games_df_3.drop_duplicates(subset=[COL_APP]).reset_index(drop=True)

# -------------------------
# Parse rank column into numeric features
# -------------------------
rank_parsed = games_df_3[COL_RANK].apply(parse_rank)
games_df_3["rank_pos"] = rank_parsed.apply(lambda x: x[0])
games_df_3["rank_total"] = rank_parsed.apply(lambda x: x[1])

# Optional: normalize rank position into a score (higher is better)
# If rank_pos=1 and rank_total=4512 -> score close to 1.0
games_df_3["rank_norm"] = 1 - (games_df_3["rank_pos"] - 1) / (games_df_3["rank_total"] - 1)
games_df_3.loc[(games_df_3["rank_total"].isna()) | (games_df_3["rank_total"] <= 1), "rank_norm"] = 0
games_df_3["rank_pos"] = games_df_3["rank_pos"].fillna(0)
games_df_3["rank_norm"] = games_df_3["rank_norm"].fillna(0)
games_df_3["rank_total"] = games_df_3["rank_total"].fillna(0)
# -------------------------
# Price buckets (recommended for content-based)
# -------------------------
games_df_3["price_bucket"] = games_df_3[COL_PRICE].apply(price_bucket)

# -------------------------
# Build combined text field
# -------------------------
games_df_3["text"] = (
    games_df_3[COL_APP].apply(clean_text) + " " +
    games_df_3[COL_CAT].apply(clean_text) + " " +
    games_df_3[COL_GENRE].apply(clean_text) + " " +
    games_df_3[COL_DESC].apply(clean_text) + " " +
    games_df_3[COL_PUBLISHER].apply(clean_text) + " " +
    games_df_3[COL_ISRANKED].apply(clean_text) + " " +
    games_df_3["price_bucket"].apply(clean_text)
)

# Quick check
# games_df_3[["name", "price", "price_bucket", "rank", "rank_pos", "rank_total", "rank_norm", "text"]].head()

display(games_df_3.head())
display(games_df_3.info())
```

This code prepares the dataset for a content-based recommendation system.
It first defines and validates required columns, ensuring important fields like name and tags exist.
Then it cleans and standardizes text fields by handling missing values, converting to strings, and normalizing formatting.
It removes duplicate games based on their name to avoid repeated items in the recommendation results.
The code also extracts structured information from the rank column and creates a price bucket feature from the numeric price.
Finally, it combines multiple fields (name, category, tags, synopsis, publisher, rank type, and price bucket) into a single text column that will be used for similarity computation.

```python
# Helper: parse genres/tags for evaluation
import ast
import re

def parse_tags(x):
    if pd.isna(x):
        return set()

    x = str(x).strip()

    if x == "no tags":
        return set()

    # Case 1: looks like a Python list stored as string
    if x.startswith("[") and x.endswith("]"):
        try:
            parsed = ast.literal_eval(x)
            return set(str(i).strip().lower() for i in parsed if str(i).strip())
        except:
            pass

    # Case 2: normal delimiter-based string
    parts = re.split(r"[;,|/]", x)
    parts = [p.strip().lower() for p in parts if p.strip()]
    return set(parts)

games_df_4 = games_df_3.copy()
games_df_4["tags"] = games_df_4[COL_GENRE].apply(parse_tags)
games_df_4[["tags"]].head()
```

This step extracts and processes genre information into structured tag sets for each application. The tags are cleaned and split based on common separators to enable comparison between items. These tag sets are later used to evaluate recommendation quality using genre overlap as a relevance proxy.

```python
# === MODEL 1 — TF-IDF + Cosine Similarity ===

# Build TF-IDF matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(
    ngram_range=(1,2),
    stop_words="english",
    min_df=2,
    max_df=0.9
)

X_tfidf = tfidf.fit_transform(games_df_4["text"])
display(X_tfidf.shape)
display(len(tfidf.vocabulary_))
```

This step transforms the combined text data into numerical feature vectors using the TF-IDF (Term Frequency–Inverse Document Frequency) method. TF-IDF represents each application based on the importance of words in its textual content.

The TF-IDF matrix shape (13145, 17027) means that 13,145 games were transformed into vectors with 17027 features each. Each row represents one game, and each column represents a word or two-word phrase (ngram). The number 17027 confirms that the vocabulary size is exactly 17027.

min_df=2 removes words that appear in fewer than 2 documents, filtering out very rare or noisy terms that don’t help generalize similarity.

max_df=0.9 removes words that appear in more than 90% of documents, eliminating overly common terms that don’t help distinguish between games.

```python
# Recommend Top-N similar apps (TF-IDF)
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

games_df_5 = games_df_4.copy()

# Map app name -> index (after reset_index, this matches X_tfidf row order)
games_df_5 = games_df_5.reset_index(drop=True)
name_to_idx = pd.Series(games_df_5.index, index=games_df_5[COL_APP]).to_dict()

# Choose which columns to show in recommendation output
OUTPUT_COLS = [
    COL_APP,        # name
    COL_GENRE,      # tags
    COL_DESC,       # sinopsis
    COL_CAT,        # Category
    COL_PRICE,      # price
    COL_RATING,     # rankScore
    COL_RANK,       # rank (e.g., "16 / 4512")
    COL_ISRANKED,   # rank type
    COL_PUBLISHER   # Published by
]

# Keep only columns that actually exist (prevents KeyError)
OUTPUT_COLS = [c for c in OUTPUT_COLS if c in games_df_5.columns]

def recommend_tfidf(app_name: str, top_n: int = 10, return_cols=None):
    if app_name not in name_to_idx:
        # helpful suggestion: show close matches
        matches = games_df_5[COL_APP].str.contains(app_name, case=False, na=False)
        suggestions = games_df_5.loc[matches, COL_APP].head(10).tolist()
        raise ValueError(f"App name not found. Try exact match. Suggestions: {suggestions}")

    idx = name_to_idx[app_name]

    # Similarity of this game against all games
    sims = cosine_similarity(X_tfidf[idx], X_tfidf).ravel()
    sims[idx] = -1  # exclude itself

    top_idx = np.argsort(sims)[::-1][:top_n]

    cols = OUTPUT_COLS if return_cols is None else return_cols
    cols = [c for c in cols if c in games_df_5.columns]

    recs = games_df_5.loc[top_idx, cols].copy()
    recs["score"] = sims[top_idx]

    # optional: nicer ordering (highest score first)
    recs = recs.sort_values("score", ascending=False).reset_index(drop=True)
    return recs

# Example
recommend_tfidf(games_df_5[COL_APP].iloc[0], top_n=10)
```

This step computes cosine similarity between the selected app and all other apps using the TF-IDF feature matrix. The apps are then ranked based on similarity scores. Finally, the top-N most similar applications are returned as recommendations.

```python
# === MODEL 2 — Sentence-BERT Embeddings + Cosine Similarity ===

from sentence_transformers import SentenceTransformer
import numpy as np

games_df_6 = games_df_4.copy().reset_index(drop=True)

# Load model (fast + good balance)
sbert_model = SentenceTransformer("all-MiniLM-L6-v2")

# Encode text column
embeddings = sbert_model.encode(
    games_df_6["text"].tolist(),
    normalize_embeddings=True,      # important for cosine similarity
    show_progress_bar=True
)

embeddings.shape
```

This step converts the combined text of each application into dense vector representations using a pre-trained Sentence-BERT model. These embeddings capture semantic meaning beyond simple word frequency. The resulting vectors are used to compute similarity between apps at a deeper contextual level.

```python
# Recommend Top-N similar apps (Embeddings)
import numpy as np

# Make sure index alignment matches embeddings
games_df_6 = games_df_4.copy().reset_index(drop=True)

# Map name -> index (aligned with embeddings)
name_to_idx_sbert = pd.Series(
    games_df_6.index,
    index=games_df_6[COL_APP]
).to_dict()

# Columns to return (reuse your OUTPUT_COLS logic if defined)
OUTPUT_COLS_SBERT = [
    COL_APP,
    COL_GENRE,
    COL_DESC,
    COL_CAT,
    COL_PRICE,
    COL_RATING,
    COL_RANK,
    COL_ISRANKED,
    COL_PUBLISHER
]

OUTPUT_COLS_SBERT = [c for c in OUTPUT_COLS_SBERT if c in games_df_6.columns]


def recommend_sbert(app_name: str, top_n: int = 10):
    if app_name not in name_to_idx_sbert:
        matches = games_df_6[COL_APP].str.contains(app_name, case=False, na=False)
        suggestions = games_df_6.loc[matches, COL_APP].head(10).tolist()
        raise ValueError(f"App name not found. Suggestions: {suggestions}")

    idx = name_to_idx_sbert[app_name]

    # Since embeddings were normalized, dot product = cosine similarity
    sims = embeddings @ embeddings[idx]

    sims[idx] = -1  # exclude itself

    top_idx = np.argsort(sims)[::-1][:top_n]

    recs = games_df_6.loc[top_idx, OUTPUT_COLS_SBERT].copy()
    recs["score"] = sims[top_idx]

    return recs.sort_values("score", ascending=False).reset_index(drop=True)


# Example
recommend_sbert(games_df_6[COL_APP].iloc[0], top_n=10)
```

This step calculates cosine similarity between the selected app’s embedding vector and all other app embeddings. The apps are then ranked according to their semantic similarity scores. The top-N most similar applications are returned as recommendations.

```python
# Because we don’t have user interactions, we define “relevant” as:
#   Recommended game shares at least 1 genre tag with the query game.

# Offline Evaluation (Precision@K with Genre Overlap Proxy)
def precision_at_k_from_names(rec_names, query_tags, df_ref, name_to_idx_ref, k=10):
    rec_names = rec_names[:k]
    if len(rec_names) == 0:
        return 0.0

    hits = 0
    for name in rec_names:
        j = name_to_idx_ref.get(name, None)
        if j is None:
            continue

        rec_tags = df_ref.loc[j, "tags"]
        if len(query_tags.intersection(rec_tags)) > 0:
            hits += 1

    return hits / k

def eval_precision_model(
    recommender_fn,
    df_ref,
    name_to_idx_ref,
    sample_size=300,
    k=10,
    seed=42
):
    rng = np.random.default_rng(seed)
    idxs = rng.choice(len(df_ref), size=min(sample_size, len(df_ref)), replace=False)

    scores = []
    for idx in idxs:
        app = df_ref.loc[idx, COL_APP]
        query_tags = df_ref.loc[idx, "tags"]

        if len(query_tags) == 0:
            continue

        rec_df = recommender_fn(app, top_n=k)
        rec_names = rec_df[COL_APP].tolist()

        p = precision_at_k_from_names(
            rec_names,
            query_tags,
            df_ref,
            name_to_idx_ref,
            k=k
        )

        scores.append(p)

    return float(np.mean(scores)) if scores else 0.0

p_tfidf = eval_precision_model(
    recommend_tfidf,
    games_df_5,
    name_to_idx,
    sample_size=300,
    k=10
)

p_sbert = eval_precision_model(
    recommend_sbert,
    games_df_6,
    name_to_idx_sbert,
    sample_size=300,
    k=10
)

p_tfidf, p_sbert
```

This code evaluates the recommendation models using an offline metric called Precision@K. Since we do not have real user interaction data, a recommendation is considered “relevant” if it shares at least one genre tag with the query game. The precision_at_k_from_names function computes how many of the Top-K recommended games overlap in genre with the query game. The eval_precision_model function samples random games, generates recommendations, and calculates the average Precision@K score. This provides a quantitative way to compare TF-IDF and SBERT models. The final results indicate that SBERT slightly outperforms TF-IDF in genre-based relevance under this proxy evaluation.

```python
# Compare results + inspect outputs
print(f"TF-IDF Precision@10 (genre overlap proxy): {p_tfidf:.4f}")
print(f"SBERT  Precision@10 (genre overlap proxy): {p_sbert:.4f}")

query_app = "Xenoblade Chronicles 3"
print("\nQuery:", query_app)

# TF-IDF (Model 1)
print("\nTF-IDF Recommendations:")
try:
    display(recommend_tfidf(query_app, top_n=10))
except ValueError as e:
    print("TF-IDF:", e)
    print("Example names:", games_df_5[COL_APP].head(20).tolist())

# SBERT (Model 2)
print("\nSBERT Recommendations:")
try:
    display(recommend_sbert(query_app, top_n=10))
except ValueError as e:
    print("SBERT:", e)
    print("Example names:", games_df_6[COL_APP].head(20).tolist())
```
